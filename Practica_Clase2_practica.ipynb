**Problema de Mezcla de Ingredientes**

```python
import numpy as np

# array
ingredientes = np.array([
    [2, 1, 0.5],    # Integral
    [1.5, 0.5, 0.3],  # Centeno
    [3, 1, 0.8]    # Blanco
])

# panes de cada tipo
cantidades = np.array([200, 150, 300])

# cuantos ingredientes en total
totales = np.dot(ingredientes.T, cantidades)

# resultados
print(f"Harina: {totales[0]}")
print(f"Agua: {totales[1]}")
print(f"Levadura: {totales[2]}")

```

**Problema de Inversión**

```python
import numpy as np

# array
A = np.array([[1, 1], [0.03, 0.05]])
b = np.array([10000, 450])

# calc inversion por cuenta
x = np.linalg.solve(A, b)
cA = round(x[0])
cB = round(x[1])

# resultados
print("Cuenta A:", cA)
print("Cuenta B:", cB)
```

**Problema de Precios de Productos**

```python
import numpy as np

# array
A = np.array([[1, 1, 1], [400, 800, 200]])
b = np.array([150, 50000])

# resolver con minimos cuadrados
x, residuals, _, _ = np.linalg.lstsq(A, b, rcond=None)

# resultados
print("Televisores:", round(x[0]))
print("Computadoras:", round(x[1]))
print("Camaras:", round(x[2]))
```

**Problema de Distribución de Productos**

```python
import numpy as np

# array costos
Costos = np.array([[5, 7, 6, 8], #Almacen A
                   [6, 5, 7, 6], #Almacen B
                   [7, 8, 5, 6]]) #Almacen C

Demandas = np.array([20, 30, 25, 15]) # demanda por tienda; suponiendo que tienda 1= 20, tienda 2= 30, tienda 4= 25 y tienda5= 15
Oferta = np.array([30, 25, 35]) # oferta almacenes; suponiendo A= 30, B= 25 y C= 35

# crear matriz A y vector b para el sistema de ecuaciones
A = np.vstack([Costos, np.identity(4)]) #ref: https://numpy.org/doc/stable/reference/generated/numpy.vstack.html
b = np.hstack([Oferta, Demandas]) #ref: https://numpy.org/doc/stable/reference/generated/numpy.hstack.html

# Resolver el sistema de ecuaciones utilizando minimos cuadrados
solucion, residuals, _, _ = np.linalg.lstsq(A, b, rcond=None)

print("Solucion:", solucion)

```

**Predicción de Ventas de Productos**

**Estimación de Tiempo de Entrega de Pedidos**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# array
distancia = np.array([2.5, 3.0, 1.8, 2.2, 3.5])
tiempo_entrega = np.array([20, 25, 18, 21, 30])

# funcion lineal
def func(X, a, b):
    return a + b*X

# curve_fit
params, covariance = curve_fit(func, distancia, tiempo_entrega)

# print
print('Intercepto (tiempo base):', params[0])
print('Tiempo por milla:', params[1])

# grafica
plt.scatter(distancia, tiempo_entrega, label='Datos reales', color='blue')
plt.plot(distancia, func(distancia, *params), 'r-', label='Ajuste lineal')
plt.title('Distancia vs Tiempo de Entrega')
plt.xlabel('Distancia (millas)')
plt.ylabel('Tiempo de Entrega (minutos)')
plt.legend()
plt.show()
```

# Predicción de Rendimiento de Estudiantes

```python
import numpy as np
import matplotlib.pyplot as plt

# array
X = np.array([5, 7, 8, 6, 9])
y = np.array([75, 85, 92, 78, 94])

# regresion lineal
m, c = np.polyfit(X, y, 1)

# funcion de la linea
def line(x, m, c):
    return m * x + c

# grafica
plt.scatter(X, y, color='blue', label='Datos reales')
plt.plot(X, line(X, m, c), 'r-', label=f'Ajuste lineal: y={m:.2f}x + {c:.2f}')
plt.xlabel('Tiempo de estudio semanal (horas)')
plt.ylabel('Calificaciones finales')
plt.legend()
plt.title('Relacion tiempo de estudio y calificaciones')
plt.show()
```

**Modelo de Crecimiento Exponencial**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# array
x1 = np.array([0, 1, 2, 3, 4])
y1 = np.array([1, 2.7, 7.4, 20.1, 54.6])

# funcion exponencial
def exponential(x, a, b):
    return a * np.exp(b * x)

params1, _ = curve_fit(exponential, x1, y1)

# grafica
plt.scatter(x1, y1, color='blue')
plt.plot(x1, exponential(x1, *params1), 'r-')
plt.title('Modelo de Crecimiento Exponencial')
plt.show()
```

**Modelo de Regresión Logística**

```python
# array
x2 = np.array([1, 2, 3, 4, 5, 6, 7])
y2 = np.array([0.047, 0.119, 0.268, 0.5, 0.731, 0.881, 0.953])

# funcion logistica
def logistic(x, a, b):
    return 1 / (1 + a * np.exp(-b * x))

params2, _ = curve_fit(logistic, x2, y2)

# grafica
plt.scatter(x2, y2, color='blue')
plt.plot(x2, logistic(x2, *params2), 'r-')
plt.title('Modelo de Regresion Logistica')
plt.show()
```

**Modelo Polinomial**

```python
# array
x3 = np.array([0, 1, 2, 3, 4])
y3 = np.array([4, 1, 0, 1, 4])

# polinomio
params3 = np.polyfit(x3, y3, 2)
poly = np.poly1d(params3)

# grafica
plt.scatter(x3, y3, color='blue')
plt.plot(x3, poly(x3), 'r-')
plt.title('Modelo Polinomial')
plt.show()
```

**Modelo de Ley de Potencias**

```python
# array
x4 = np.array([1, 2, 3, 4, 5])
y4 = np.array([3, 8, 15, 24, 35])

# ley de potencias
def ley_powr(x, a, b):
    return a * x**b

params4, _ = curve_fit(ley_powr, x4, y4)

# grafica
plt.scatter(x4, y4, color='blue')
plt.plot(x4, ley_powr(x4, *params4), 'r-')
plt.title('Modelo de Ley de Potencias')
plt.show()
```
