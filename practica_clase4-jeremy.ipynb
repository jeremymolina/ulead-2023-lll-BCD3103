**Ejercicio #1**

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt


#Definir la funcion de la ODE
def modelo_demanda(y, t):
  dydt = 0.1 * y
  return dydt

# Condiciones inciales
y0 = 1000 #Demanda inicial

t = np.linspace(0, 6, 10) # Un dia


# Resolver EDO
sol = odeint(modelo_demanda, y0, t)

#graficar
plt.plot(t, sol)
plt.xlabel('Crecimiento exponencial en horas')
plt.ylabel('Cantidad de Bacterias')
plt.show()
```

**Ejercicio #2**

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def enfriamiento(T, t, k):
    dTdt = -k * (T - 20)
    return dTdt

# Defined values
T0 = 100

# Time
t = np.linspace(0, 2, 100)  # 2 horas

# K, random
k = 0.02 #muestreo

# Solve ODE
sol = odeint(enfriamiento, T0, t, args=(k,))

# Chart
plt.plot(t, sol)
plt.xlabel('Tiempo en horas')
plt.ylabel('Temperatura Celsius')
plt.title('Enfriamiento del Objecto')
plt.show()

# Print result
print(f"Temperatura despues de 2 horas: {sol[-1][0]:.2f}°C")
```

**Ejercicio 3**

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Definamos formula
def mas_formula(y, t, frec):
    x, v = y
    dxdt = v # Esto se obtiene al despejar v que es igual a dxdt
    dvdt = -frec**2 * x # Esto se obtiene al despejar −ω^2x
    return [dxdt, dvdt] #retornamos los valores para la grafica

# Frecuencia Angular
frec = 2.0  # muestreo

# Initial conditions
x = 1.0 # posicion inicial
v = 0.0  # velocidad inicial

# Time
t = np.linspace(0, 10, 1000)  #  10 seconds

# Solve ODE
sol = odeint(mas_formula, [x, v], t, args=(frec,))

# Get from mas_formula the position and velocity
x, v = sol[:, 0], sol[:, 1]

# Chart
plt.figure()
plt.plot(t, x, label='Posicion (x)')
plt.plot(t, v, label='Velocidad (v)')
plt.xlabel('Tiempo')
plt.ylabel('Valor')
plt.title('Frecuencia Angular')
plt.legend()
plt.show()
```
